```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('DATA/fake_reg.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['feature1', 'feature2']].values

y = df['price'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

```python
# normalize the data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
# import packages from keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
 help(Sequential)
```

```python
# create a model Option 1
model = Sequential([Dense(4,activation='relu'),
                   Dense(2, activation='relu'),
                   Dense(1)])
```

```python
model = Sequential()

model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))

# Final output node for prediction
model.add(Dense(1))

model.compile(optimizer='rmsprop',loss='mse')
```

### Choosing an optimizer and loss

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

```python
model.compile(optimizer='rmsprop', loss='mse')
```

```python
# deploy the model
model.fit(x_train, y_train, epochs=250)
```

```python
# plot the loss graph
model_history = pd.DataFrame(model.history.history)
model_history.plot()
```

```python
# get the loss on the test data
model.evaluate(x_test, y_test, verbose=0)
```

```python
# get the loss on the train data
model.evaluate(x_train, y_train, verbose=0)
```

```python
# predict the test instances
predictions = model.predict(x_test)
```

```python
# turn predictions into a data frame
test_pred = pd.DataFrame(predictions.reshape(300,))
```

```python
# create a data frame with the correct predictions
pred_df = pd.DataFrame(y_test, columns=['True Y'])
```

```python
# concatenate my predictions against the real predictions
pred_df = pd.concat([test_pred, pred_df], axis = 1)
```

```python
# set column names
pred_df.columns = ['Predicted Y', 'True Y']
```

```python
# visualize my predictions against the real predictions
pred_df
```

```python
# plot my predictions against the real predictions
sns.scatterplot(x = 'Predicted Y', y = 'True Y', data = pred_df)
```

```python
# import regression metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# print the usual metrics to evaluate the model
print('MAE: ', mean_absolute_error(pred_df['True Y'], pred_df['Predicted Y']))
print('MSE: ', mean_squared_error(pred_df['True Y'], pred_df['Predicted Y']))
print('RMSE: ', np.sqrt(mean_squared_error(pred_df['True Y'], pred_df['Predicted Y'])))
```
