```python
# imports
import pandas as pd
import numpy as np
```

```python
# viz imports
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```python
# use sklearn to generate artificial data
from sklearn.datasets import make_blobs
```

```python
# generate aritifical data
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
# check the shape of the data set
data[0].shape
```

```python
# see our clusters
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
```

```python
# import k means from sklearn
from sklearn.cluster import KMeans
```

```python
# create our algorithm
kmeans = KMeans(n_clusters=4)
```

```python
# fit the algorithm
kmeans.fit(data[0])
```

```python
# get the clusters center
kmeans.cluster_centers_
```

```python
# get the cluster for each record
kmeans.labels_
```

```python
# plot the kmeans output vs the real data
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize = (10, 6))

ax1.set_title('Kmeans output')
ax1.scatter(data[0][:, 0], data[0][:,1], c = kmeans.labels_, cmap= 'rainbow')

ax2.set_title('Real output')
ax2.scatter(data[0][:, 0], data[0][:,1], c = data[1], cmap= 'rainbow')
```
