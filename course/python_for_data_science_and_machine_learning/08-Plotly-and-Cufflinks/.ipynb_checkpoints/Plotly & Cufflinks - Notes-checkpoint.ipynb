### Plotly & Cufflinks

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# check the version of plotly
from plotly import __version__
```

```python
# import cufflinks
import cufflinks as cf
```

```python
# import useful packages of plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# connect to Java Script
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
# create a DF
df = pd.DataFrame(np.random.randn(100, 4), columns='a b c d'.split())
```

```python
# create a DF2
df2 = pd.DataFrame({'Categories':['A', 'B', 'C'], 'Values': [32, 43, 50]})
```

```python
# create an interactive plot with iplot
df.iplot()
```

```python
# create an interactive scatter plot
df.iplot(kind='scatter', x='a', y='b', mode='markers')
```

```python
# create an interactive bar plot
df2.iplot(kind='bar', x='Categories', y='Values')
```

```python
# create a box plot
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y':[10,20,30,20,10], 'z':[5,4,3,2,1]})
```

```python
# create a surface plot 3D
df3.iplot(kind='surface', colorscale='rdylbu')
```

```python
# create an interactive histogram
df.iplot(kind='hist', bins=10)
```

```python
# create a spread splot > very useful for financial data
df[['a', 'b']].iplot(kind='spread')
```

```python
# bubble plot
df.iplot(kind='bubble', x='a', y='b', size ='c')
```

```python
# create a scatter matrix plot
df.scatter_matrix()
```
