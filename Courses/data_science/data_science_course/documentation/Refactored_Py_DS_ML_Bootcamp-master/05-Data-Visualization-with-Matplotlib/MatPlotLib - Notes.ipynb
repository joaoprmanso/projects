```python
import matplotlib.pyplot as plt
```

```python
# allow us to visualize the graphics inside jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 3
```

```python
# Functional way of ploting
plt.plot(x, y,)
plt.xlabel('Day')
plt.ylabel('New Confirmed Cases')
plt.title('Covid-19 Growth')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 1, 1])

axes.plot(x, y)
axes.set_xlabel('Day')
axes.set_ylabel('New cases')
axes.set_title('My First Chart')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.95, 0.95])
axes2 = fig.add_axes([0.2, 0.2, 0.8, 0.8])
axes1.plot(x, y)
axes2.plot(x, y)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y)
```

```python
# second method to create plots
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

# axis 0
axes[0].plot(x, y)
axes[0].set_title('Axis 0')
# axis 1
axes[1].plot(x, y)
axes[1].set_title('Axis 1')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

axes[0].plot(x, y)
axes[0].set_title('Last example 0')

axes[1].plot(x, y)
axes[1].set_title('Last example 1')

axes[2].plot(x, y)
axes[2].set_title('Last example 2')

plt.tight_layout()
```

### Figure Size and DPI

```python
# set the size
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(6, 3))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.jpeg')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x, label='Scenario 1')
ax.plot(x, x**2, label='Scenario 2')
ax.plot(x, x**3, label='Scenario 3')

ax.set_title('kjne')
ax.set_ylabel('dejn')
ax.set_xlabel('diejn')

ax.legend(loc=0)
```

### Matplotlib - Part 3 - Plot Asthetics

```python
# create a new figure
fig = plt.figure()

# add the axes to the new figure
axes = fig.add_axes([0,0,1,1])

# set the color, line width, alpha and line style
axes.plot(x, y, color='black',lw=3, alpha=0.5, ls='-', marker='o', markersize=2,
         markerfacecolor='blue', markeredgewidth=3, markeredgecolor='green')

axes.set_xlim([0, 1])
axes.set_ylim([0, 4])
```
